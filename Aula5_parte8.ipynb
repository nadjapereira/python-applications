```python
import simplejson as json
import pandas as pd
```

```python
dados = []
with open('tweets_5min.json') as arquivo:
    for linha in arquivo:
        dados.append(json.loads(linha))
```

```python
df = pd.DataFrame(dados)
```

```python
print("--> Dados do tweet")
print(df['text'][0])
print(df['created_at'][0])
print(df['coordinates'][0])
print(df['retweet_count'][0])

print("\n--> Dados do usuário")
print(df['user'][0]['screen_name'])
print(df['user'][0]['location'])
print(df['user'][0]['lang'])
print(df['user'][0]['followers_count'])
```

```python
from geopy.geocoders import Nominatim
def pegar_lat_long(local):
    geo = Nominatim()
    try:
        local = geo.geocode(local)
        return (local.latitude, local.longitude)
    except:
        return 0
```

```python
print(pegar_lat_long('Av. Paulista, 302, São Paulo'))
```

```python
def salvar_hashtag(texto):
    palavras = texto.split()
    aux = []
    for p in palavras:
        if p.startswith('#'):
            aux.append(p)
    return ' '.join(aux)
```

```python
print(salvar_hashtag(df['text'][0]))
```

```python
# versao de salvar_hashtag em compressao de lista
def salvar_hashtag2(texto):
    return ' '.join([x for x in texto.split() if x.startswith('#')])
```

```python
print(salvar_hashtag2(df['text'][0]))
```

```python
colunas = ['text', 'created_at', 'coordinates', 'retweet_count', 'screen_name', 
           'location', 'lang', 'followers_count', 'lat', 'long', 'hashtags']
df_final = pd.DataFrame(columns=colunas)
df_final
```

```python
import time
t0 = time.time()
for i in range(0, len(df)):
    if df['user'].iloc[i]['location'] != None:
        tem_lat_long = pegar_lat_long(df['user'].iloc[i]['location'])
        if tem_lat_long != 0:
            df_final.loc[i] = [df['text'].iloc[i],
                               df['created_at'].iloc[i],
                               df['coordinates'].iloc[i],
                               df['retweet_count'].iloc[i],
                               df['user'].iloc[i]['screen_name'],
                               df['user'].iloc[i]['location'],
                               df['user'].iloc[i]['lang'],
                               df['user'].iloc[i]['followers_count'],
                               tem_lat_long[0],
                               tem_lat_long[1],
                               salvar_hashtag(df['text'].iloc[i]),
                              ]
            print(i, end = ' ')
tf = time.time() - t0
print("Demorou {} minutos".format(round(tf/60, 2)))
```

```python
df_final.reset_index(inplace=True, drop=True)
df_final
```

```python
df_final.to_csv('saida.csv', sep=';', index=False, encoding='utf-8')
```
